```python
import pandas as pd
import time
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By

# 봇 감지를 피하기 위해 undetected_chromedriver 사용
driver = uc.Chrome()

try:
    # 1. 일단 메인 페이지나 기록실 페이지로 접속 시도
    driver.get("https://www.koreabaseball.com/Record/Player/BatterBasic/Basic1.aspx")
    
    print("★ [수동 조작 단계] ★")
    print("1. 브라우저 창에서 에러 페이지가 뜬다면 '홈으로 가기' 등을 눌러 정상 페이지로 이동하세요.")
    print("2. '기록실' 메뉴에서 'LG' 팀을 선택해 표가 나오게 하세요.")
    print("3. 화면에 LG 선수들 표가 확실히 보이면, VS Code 터미널에서 Enter를 누르세요.")
    
    input("LG 선수 표가 보이면 엔터를 누르세요...") # 사용자가 조작할 시간을 벌어줍니다.

    # 2. 현재 화면에 보이는 모든 HTML 소스를 가져옵니다.
    # 특정 요소를 찾지 않고 전체 소스를 긁는 방식이 에러가 훨씬 적습니다.
    html_source = driver.page_source
    
    # 3. Pandas로 표 데이터 추출
    df_list = pd.read_html(html_source)
    
    # 보통 첫 번째 테이블에 데이터가 있습니다.
    if df_list:
        df = df_list[0]
        print("--- 데이터 수집 성공! ---")
        print(df.head())
        
        # CSV 파일로 저장
        df.to_csv("lgtwins_official_data.csv", index=False, encoding='utf-8-sig')
    else:
        print("표를 찾지 못했습니다. 다시 시도해 보세요.")

except Exception as e:
    print(f"오류 발생: {e}")

finally:
    driver.quit()
```
